# Matplotlib
- plotting library for the Python
- numerical mathematics extension is numpy
- provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits
- high quality output
- support for custom labels and texts
- very customizable in general
- great control of every elementin a figure

```python
import matplotlib.pyplot as plt
%matplotlib inline # no need to write plt.show() after writing this line
```

```python
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(50,50)
b = np.arange(50,60)
```

```python
# plotting scatter plot

plt.scatter(x,y,c='b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
plt.savefig('test.png')
```

```python
y = x*x
```

```python
plt.plot(x,y,'r*', linestyle='dashed',
        linewidth=2,markersize=12)
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Graph in 2D')
```

```python
# creating subplots

plt.subplot(2,2,1)
plt.plot(x,y,'bo')

plt.subplot(2,2,2)
plt.plot(x,y,'g.')

plt.subplot(2,2,3)
plt.plot(x,y,'y--')
```

### Sine waves

```python
np.pi
```

```python
x = np.arange(0, 4*np.pi, 0.1)
y = np.sin(x)
plt.title('sine wave form')

plt.plot(x,y)
```

```python
# subplot
x = np.arange(0, 5*np.pi, 0.2)
y_cos = np.cos(x)
y_tan = np.tan(x)

plt.subplot(2,1,1)
plt.plot(x,y_cos,'g-')
plt.title('cose')

plt.subplot(2,1,2)
plt.plot(x,y_tan,'r-')
plt.title('tan')
```

### Barplot

```python
x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1,color='y')
plt.bar(x2,y2,color='g')
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis') 
```

### Histograms

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a) 
plt.title("histogram") 
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a,bins=20) #all range will be divided by 20 
plt.title("histogram") 
```

### Box Plot

```python
df = [np.random.normal(0,std,100) for std in range(1,4)]

plt.boxplot(df,vert=True,patch_artist=True)
```

```python
plt.boxplot(df,vert=False,patch_artist=False)
```

### Piechart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [210, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.4,0.1,0,0) # explode 1st and 2nd slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True)
plt.axis('equal')
```
