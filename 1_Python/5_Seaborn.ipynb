```python
import seaborn as sns
```

```python
df = sns.load_dataset("tips")
df.head()
```

In above dataset, tip is the dependent variable and others are the independent variable.

## Correlation with Heatmap
- uses colored cells, typically in a monochromatic scale to show a 2D correlation matrix (table) between two discrete dimensions or event types
- very important in Feature Selection

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

## JoinPlot
- allows us to study relationship between 2 numeric variables
- central chart display their correlation
- usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

## Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

## Pair Plot
- alos known as a scatterplot, in which one variable in the same data row is matched with another variable's value

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

## Dist Plot
- helps to check the distribution of the columns feature

```python
sns.distplot(df['tip']) #kde=True, KDE=kernel Density Estimation
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

## Categorical Plots
- boxplot
- violinplot
- countplot
- barplot

```python
# count plot

sns.countplot('smoker',data=df)
```

```python
sns.countplot(y='sex',data=df)
```

```python
# bar plot

sns.barplot(x='total_bill',y='time',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
df.head()
```

```python
# box plot

sns.boxplot('smoker','total_bill',data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
sns.boxplot(x='total_bill',y='day',hue='sex',data=df)
```

```python
# violin plot- helps to see both distribution of data in terms of kde and box plot

sns.violinplot(x='total_bill',y='day',data=df,palette='flare')
```
